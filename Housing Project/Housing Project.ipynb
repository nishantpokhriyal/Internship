```python
# Taking Care of null values

def missing_values(df):    
     
    from sklearn.impute import SimpleImputer
    imp_median=SimpleImputer(missing_values=np.nan, strategy='median')
    imp_mode=SimpleImputer(missing_values=np.nan, strategy='most_frequent')
    
    # checking null values
    na_col = df.isnull().sum()
    # columns which have null values
    na_col = na_col[na_col > 0]

    for i in na_col.index:
        if df[i].dtype.name == 'object':
            df[i]=imp_mode.fit_transform(df[i].values.reshape(-1,1))
            
        elif df[i].dtype.name == 'float64' or df[i].dtype.name == 'int64' or df[i].dtype.name == 'int32':
            df[i]=imp_median.fit_transform(df[i].values.reshape(-1,1))
            
    return df
```

# Removing Outliers

```python
# Removing Outliers
from numpy import percentile

a=hp.columns
for i in a:
    if isinstance(hp[i][0],str):
        continue
    else:
        # Quartiles
        q=percentile(hp[i],[25,75])
        # Lower & Upper bound
        lower_bound= q[0] - (1.5*(q[1]-q[0]))
        upper_bound= q[1] + (1.5*(q[1]-q[0]))
        hp[i] = hp[i].apply(lambda x: upper_bound if x > upper_bound else (lower_bound if x < lower_bound else x))
```

**We have removed the outliers higher than the upper whisker by the value of the upper whisker and the outliers lower than the lower whisker by the value of the lower whisker.**

```python
#List of numerical and non-numerical columns
categorical = []
numerical = []

for i in list(hp):
    if hp[i].dtypes == 'object':
        categorical.append(i)  
    else:
        numerical.append(i)
print('Categorical columns:\n',categorical)

print('Numerical columns:\n',numerical)
```

```python
# We will convert the categorical variables into numeric by applying one hot encoding
hp_dummy = pd.get_dummies(hp[categorical], drop_first=True)
```

```python
#Concating the dataframe with dummies with the the original dataframe (hp)
hp = pd.concat([hp, hp_dummy], axis=1)
hp.shape
```

# Model Creation

```python
x=hp.drop(columns=['SalePrice'])
y=hp[['SalePrice']]

print(x.shape)
print(y.shape)
```

**We have created two variables containing Input data (x) and Output data (y).**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=50)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
# Importing linear Regression algorithm, mean absolute error, mean squared error and r2 score to make the prediction model and checking its accuracy
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.metrics import r2_score
```

**Making a function to calculate and print following features for the model : mean absolute error, mean squared error, r2 score.**

```python
def results(actual,pred):
    mae=mean_absolute_error(actual,pred)
    mse=mean_squared_error(actual,pred)
    r2=r2_score(actual,pred)
    
    print('Mean Absolute Error :',mae,'\n')
    print('Mean Squared Error :',mse,'\n')
    print('r2 Score :',r2,'\n')
```

**Making a Cross Validation function for "Sale Price"**

```python
from sklearn.model_selection import cross_val_score
def accuracy(model,x_train=x_train,y_train=y_train):
    accuracies=cross_val_score(estimator=model,X = x_train, y = y_train ,cv=10)
    print('Cross_val_Accuracy: {:.2f} %'.format(accuracies.mean()*100),'\n')
    print('Cross_val_Standard Deviation: {:.2f} %'.format(accuracies.std()*100),'\n')
```

### Linear Regression Model

```python
ln=LinearRegression()
ln.fit(x_train,y_train)
pred=ln.predict(x_test)

print('\033[1m' + 'LINEAR REGRESSION MODEL SUMMERY\n' + '\033[0m')
print('Predicted Sale Price',pred)
print('\nActual Sale Price\n',y_test)
print('\n')
results(y_test,pred)
accuracy(ln)
```

### Lasso Regression

```python
ls=Lasso(alpha=0.0001)
ls.fit(x_train,y_train)
pred_ls=ls.predict(x_test)

print('\033[1m' + 'LASSO REGRESSION MODEL SUMMERY\n' + '\033[0m')
print('Predicted Sale Price',pred_ls)
print('\nActual Sale Price\n',y_test)
print('\n')
results(y_test,pred_ls)
accuracy(ls)
```

## Gradient Boosting Regressor

```python
gb=GradientBoostingRegressor()
gb.fit(x_train,y_train)
pred_gb=gb.predict(x_test)

print('\033[1m' + 'GRADIENT BOOSTING REGRESSOR MODEL SUMMERY\n' + '\033[0m')
print('Predicted Sale Price',pred_gb)
print('\nActual Sale Price\n',y_test)
print('\n')
results(y_test,pred_gb)
accuracy(gb)
```

### Random Forest Regressor

```python
rr=RandomForestRegressor()
rr.fit(x_train, y_train)
pred_rr=rr.predict(x_test)

print('\033[1m' + 'RANDOM FOREST REGRESSOR MODEL SUMMERY\n' + '\033[0m')
print('Predicted Sale Price',pred_rr)
print('\nActual Sale Price\n',y_test)
print('\n')
results(y_test,pred_rr)
accuracy(rr)
```

### We can see that the best Performing Model is Gradient Boosting Regressor Model
