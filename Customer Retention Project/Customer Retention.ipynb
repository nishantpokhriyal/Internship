## How many People will recommend each Indian retailer  to a friend

```python
z=df.iloc[:,[0,1,2,3,24]]
z
```

```python
amazon=z[z['Which of the Indian online retailer would you recommend to a friend?'].str.contains('Amazon')]
flipkart=z[z['Which of the Indian online retailer would you recommend to a friend?'].str.contains('Flipkart')]
myntra=z[z['Which of the Indian online retailer would you recommend to a friend?'].str.contains('Myntra')]
paytm=z[z['Which of the Indian online retailer would you recommend to a friend?'].str.contains('Paytm')]
snapdeal=z[z['Which of the Indian online retailer would you recommend to a friend?'].str.contains('snapdeal')]

print('Amazon :\n',amazon['1Gender of respondent'].value_counts(),amazon.shape,
      '\n\nFlipkart :\n',flipkart['1Gender of respondent'].value_counts(),flipkart.shape,
      '\n\nMyntra :\n',myntra['1Gender of respondent'].value_counts(),myntra.shape,
      '\n\nPaytm :\n',paytm['1Gender of respondent'].value_counts(),paytm.shape,
      '\n\nSnapdeal :\n',snapdeal['1Gender of respondent'].value_counts(),snapdeal.shape)
```

Amazon is the most recommended retailer by people as 219 people out of 269 would recommend it to a friend, Flipkart follows with 127 recommendations and Myntra an Paytm are recommended by 76 and 44 people while Snapdeal is the least recommended retailer with just 11 recommendations.
